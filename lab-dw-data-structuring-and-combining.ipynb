# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

```python
# Your code goes here
insurance_completo_df.info()
```

```python
insurance_completo_copy_df=insurance_completo_df.copy()
```

```python
#Poner los nombres en minusculas y quitar los espacios
insurance_completo_copy_df.columns = insurance_completo_copy_df.columns.str.lower().str.replace(" ", "_")
insurance_completo_copy_df.head()
```

```python
# Your code here
#Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", 
#which need to be standardized, for example, to "M" and "F".
insurance_completo_copy_df.nunique()
```

```python
insurance_completo_copy_df.st.value_counts()
```

```python
insurance_completo_copy_df.st.unique()
```

```python

gender_mapping = {"Washington":"Washington",
                  "Arizona":"Arizona",
                  "Nevada":"Nevada",
                  "California":"California",
                  "Oregon":"Oregon",
                  "Cali":"California",
                  "AZ":"Arizona",
                  "WA":"Washington",
                  "nan":"nan"}

insurance_completo_copy_df['st'] = insurance_completo_copy_df['st'].map(gender_mapping)
insurance_completo_copy_df.head()   
```

```python
insurance_completo_copy_df.education.unique()
```

```python
education_mapping = {"Master":"Master",
                  "Bachelor":"Bachelor",
                  "High School or Below":"High School or Below",
                  "College":"College",
                  "Bachelors":"Bachelor",
                  "Doctor":"Doctor",
                  "nan":"nan"}

insurance_completo_copy_df['education'] = insurance_completo_copy_df['education'].map(education_mapping)
insurance_completo_copy_df.head()   
```

```python
insurance_completo_copy_df.education.unique()
```

```python
insurance_completo_copy_df.info()
```

```python
insurance_completo_copy_df.vehicle_class.unique()
```

```python

car_mapping = {"Four-Door Car":"Four-Door Car",
                  "Two-Door Car":"Two-Door Car",
                  "SUV":"SUV",
                  "Luxury SUV":"Luxury",
                  "Sports Car":"Luxury",
                  "Luxury Car":"Luxury",
                  "nan":"nan"}
insurance_completo_copy_df['vehicle_class'] = insurance_completo_copy_df['vehicle_class'].map(car_mapping)
insurance_completo_copy_df.head()   
```

```python
insurance_completo_copy_df.vehicle_class.unique()
```

```python
insurance_completo_copy_df.customer_lifetime_value.unique()
completo_2_df = insurance_completo_copy_df.copy()
```

```python
completo_2_df['customer_lifetime_value'] = completo_2_df['customer_lifetime_value'].apply(lambda x: x.replace('%', '') if isinstance(x, str) and x.endswith('%') else x)
completo_2_df
```

```python
completo_2_df['number_of_open_complaints'] = completo_2_df['number_of_open_complaints'].apply(lambda x: x.split('/')[1] if isinstance(x, str) else x)
completo_2_df.head()
```

```python
completo_2_df.number_of_open_complaints.unique()
```

```python
completo_3_df=completo_2_df.copy()
completo_3_df.isna().sum()
```

```python
len(completo_3_df)
```

```python
completo_3_df["customer_lifetime_value"] = completo_3_df["customer_lifetime_value"].astype(float)
completo_3_df.info()
```

```python
completo_3_df["number_of_open_complaints"] = completo_3_df["number_of_open_complaints"].astype(float)
completo_3_df.info()
```

```python
completo_4_df=completo_3_df.copy()
completo_4_df.isna().sum()
```

```python
len(completo_4_df)
```

```python
round(completo_4_df["st"].isna().sum() / len(completo_4_df) * 100, 1)
```

```python
completo_4_df = completo_4_df.dropna(how="all")
len(completo_4_df)
```

```python
completo_4_df["gender"].bfill(inplace=True)
completo_4_df.isna().sum()
```

```python
completo_4_df["customer_lifetime_value"] = completo_4_df["customer_lifetime_value"].fillna(completo_4_df["customer_lifetime_value"].mean())
```
